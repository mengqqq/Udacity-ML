**任务3.4**：分析1968年~2015年六月电影的数量的变化


```python
# 筛选条件1 年份        
# 筛选条件2 六月

emd["release_date"]=pd.to_datetime(emd["release_date"])
emd["release_month"]=emd["release_date"].dt.month
emd["release_year"]=emd["release_date"].dt.year

year=pd.Series(emd["release_year"])
sel_year=year.between(1968,2015)
month=pd.Series(emd["release_month"])
sel_June=month.between(5.5,6.5)
#sel_June = None
# 筛选数据并作图(参考逻辑读取部分)
emd[sel_year&sel_June]['release_year'].value_counts().sort_index().plot(kind='bar', figsize=(20, 10))
plt.xlabel("release_year")
plt.ylabel("sel_year&sel_June")
plt.title("1968-2015 movies numbers changing")
# TODO: 补充可视化的其他要素
```

```python
**任务3.5**：分析1968年~2015年六月电影 Comedy 和 Drama 两类电影的数量的变化。
```

```python
sel_comedy=emd["genres"].str.contains("Comedy")
sel_drama=emd["genres"].str.contains("Drama")
emd[sel_year&sel_June&sel_comedy]["release_year"].value_counts().sort_index().plot(kind="line",figsize=(10,5),label="Comedy")
```

```python
emd[sel_year&sel_June&sel_drama]["release_year"].value_counts().sort_index().plot(kind="line",figsize=(10,5),label="Drama")
```
